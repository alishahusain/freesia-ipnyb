```python
import pandas as pd
import numpy as np
```

```python
# data input

node_data = pd.read_csv("Standardized Outputs/node2-3-17-18-7-8combined.csv")
node_data.columns
```

```python

node_column = node_data.columns[13:]

node_values = node_data[node_column].values

```

```python

node_values = np.asarray(node_values)
node_values=np.swapaxes(node_values, 0, 1)

```

```python
intersection_matrix = []

for j in node_values:
    intersection_list=[]
    for k in node_values:
        count = 0
        i = 0
        while i <= 552:
            if j[i] == 1:
                if j[i] == k[i]:
                    count = count + 1
            i = i+1
        intersection_list.append(count)
    intersection_matrix.append(intersection_list)
print(intersection_matrix)
```

```python
col = [0,1,2,6,7,9,11,13,17,21,23,24,27,72,74]
df = pd.DataFrame(intersection_matrix, columns=col, index=col)
```
