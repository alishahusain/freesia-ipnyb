```python
import pandas as pd
```

```python
indust = pd.read_csv("ACSDT5Y2018.C24030_data_with_overlays_2020-07-22T172737.csv")
```

```python
indust.fillna(0)
cols = indust.columns.drop(['GEO_ID', 'NAME'])
indust[cols] = indust[cols].apply(pd.to_numeric, errors='coerce')
```

```python
indust
```

```python
# Popular Industries----
# men finance: C24030_014E
# womens finance: C24030_041E
# male healthcare: C24030_021E
# womens healthcare: C24030_048E
# male retail: C24030_009E
# female retail: C24030_036E
# male food: C24030_024E
# female food: C24030_051E
# men scientific: C24030_051E
# female scientific: C24030_044E
######
# All other industries----
# male agri: C24030_003E
# male construction: C24030_006E
# male manufacturing: C24030_007E
# male wholesale trade: C24030_008E
# male transpo: C24030_010E
# male info: C24030_013E
# male other: C24030_027E
# male PA: C24030_028E
# female agri: C24030_030E
# female construct: C24030_033E
# female manufacturing: C24030_034E
# female wholesale trade: C24030_035E
# female transpo: C24030_037E
# female info: C24030_040E
# female other: C24030_054E
# female pa: C24030_055E



indust["pop_indust_sum"] = indust["C24030_014E"] + indust["C24030_041E"] + indust["C24030_021E"] + indust["C24030_048E"] + indust["C24030_009E"] + indust["C24030_036E"] + indust["C24030_024E"] + indust["C24030_051E"] + indust["C24030_051E"] + indust["C24030_044E"]


indust["other_indust_sum"] = indust["C24030_003E"] + indust["C24030_006E"] + indust["C24030_007E"] + indust["C24030_008E"] + indust["C24030_010E"] + indust["C24030_013E"] + indust["C24030_027E"] + indust["C24030_028E"] + indust["C24030_030E"] + indust["C24030_033E"] + indust["C24030_034E"] + indust["C24030_035E"] + indust["C24030_037E"] + indust["C24030_040E"] + indust["C24030_054E"] + indust["C24030_055E"]



indust["total"] = indust["pop_indust_sum"] + indust["other_indust_sum"]
indust["pop_indust_sum"] = indust["pop_indust_sum"]/indust["total"]
indust["other_indust_sum"] = indust["other_indust_sum"]/indust["total"]

```

```python
to_export = indust[["GEO_ID", "pop_indust_sum", "other_indust_sum"]]
to_export = to_export.iloc[1:,:]
```
